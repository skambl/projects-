```python
X = df_filled.drop('capacity_mw', axis=1)
y_capacity = df_filled['capacity_mw']
X_train, X_test, y_capacity_train, y_capacity_test = train_test_split(X, y_capacity, test_size=0.2, random_state=42)
```
